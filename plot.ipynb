```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ppo = 'PPO_'
rnd = 'RND_'
ospoe = 'OSPOE_'
dirname = 'MountainCarCsv'
```

```python
table = {}

for i in range(5):
    j = i + 1
    for name in ['PPO', 'RND', 'OSPOE']:
        if name not in table:
            table[name] = []
        csv = pd.read_csv(f'{dirname}/{name}_{j}.csv')
        table[name].append(csv['Value'].to_numpy())
```

```python
min_length = np.float('inf')
for key in table.keys():
    for arr in table[key]:
        min_length = min(min_length, len(arr))

# truncate np arrays
for key in table.keys():
    for i in range(len(table[key])):
        table[key][i] = table[key][i][:min_length]

# get means and stds
stats = {}
for key in table.keys():
    np_arr = np.array(table[key])
    stats[key] = {
        'mean' : np_arr.mean(0),
        'std' : np_arr.std(0),
    }
```

```python
plt.figure(dpi=200)
for name in ['PPO', 'RND', 'OSPOE']:
    mean = stats[name]['mean']
    std = stats[name]['std']
    plt.plot(mean, label=name)
    plt.fill_between(np.arange(min_length), mean - std, mean + std, alpha=0.2)

plt.title('Performance On Normalized MountainCar')
plt.legend()
```

```python
class normalizer():
    self.mean = 0
    self.std = 1
    
    def add(state):
        self.mean_new <- state, self.mean
        self.std_new <- state, self.std
        self.mean = self.mean_new
        self.std = self.std_new
        
    def transfrom(state):
        return (state - self.mean) / (self.std + 1e-8)
```
